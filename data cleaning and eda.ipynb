```python
iitrec=data.fillna({'Gender': 'Gender-Neutral'})
```

```python
sns.heatmap(iitrec.isnull(),yticklabels=False)
```

```python
iitrec.info()
```

```python
iitrec['Opening Rank']=iitrec['Opening Rank'].replace('[^\d.]','',regex=True).astype(int)
```

```python
iitrec['Closing Rank']=iitrec['Closing Rank'].replace('[^\d.]','',regex=True).astype(int)
```

```python
iitrec.info()
```

```python
sns.heatmap(iitrec.isnull(),yticklabels=False)
```

```python
# Find rows with non-numeric values

# Set display options to show all rows
pd.set_option('display.max_rows', None)

non_numeric_rows = iitrec[iitrec['Opening Rank'].isna()]

# Print the resulting DataFrame
print(non_numeric_rows)
```
